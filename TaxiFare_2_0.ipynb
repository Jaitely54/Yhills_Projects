# **To predict a rider's taxi fare**

*Importing Libraries*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
```

*Loading DataSet*

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets-main/TaxiFare.csv')
df
```

```python
df.info()
```

no null values

```python
df.head(20)
```

here i found that there is 0 in the columns which can lead to wrong prediction

```python
columns_mean = ['longitude_of_pickup', 'latitude_of_pickup', 'longitude_of_dropoff', 'latitude_of_dropoff']
# Calculate the mean of each column (excluding zeros)
for column in columns_mean:
    mean_Value = df[df[column] != 0][column].mean()
    df[column] = df[column].replace(0, mean_Value)
```

```python
df.head(30)
```

```python
df.sample(10)
```

```python
df.drop(['date_time_of_pickup','unique_id'],axis=1,inplace=True)
```

```python
df.info()
```

```python
# Split the data into features (X) and target (Y)
X = df.drop(columns=['amount'],axis=1)
Y = df['amount']
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
model_1 = DecisionTreeRegressor()
```

```python
model_1.fit(X_train,Y_train)
```

```python
model_1.score(X_train,Y_train)
```

```python
model_1.score(X_test,Y_test)
```

```python
model_2 = GradientBoostingRegressor()
```

```python
model_2.fit(X_train,Y_train)
```

```python
model_2.score(X_train,Y_train)
```

```python
model_2.score(X_test,Y_test)
```

```python
model_3 = RandomForestRegressor()
```

```python
model_3.fit(X_train,Y_train)
```

```python
model_3.score(X_train,Y_train)
```

```python
model_3.score(X_test,Y_test)
```

```python
model_4 = LinearRegression()
```

```python
model_4.fit(X_train,Y_train)
```

```python
model_4.score(X_train,Y_train)
```
